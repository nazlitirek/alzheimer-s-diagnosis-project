PhysicalActivity: Weekly physical activity in hours, ranging from 0 to 10.

```python
df = df.drop(['PatientID', 'DoctorInCharge'], axis=1)
```

```python
# Korelasyon matrisi
corr_matrix = df.corr(numeric_only=True)

# IsÄ± haritasÄ±
plt.figure(figsize=(18, 15))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Korelasyon Matrisi IsÄ± HaritasÄ±', fontsize=16)
plt.show()
```

```python
correlation_with_target = corr_matrix['Diagnosis'].sort_values(ascending=False)
print(correlation_with_target)
```

```python
top_corr = correlation_with_target.drop('Diagnosis').abs().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_corr.values, y=top_corr.index, palette='viridis')
plt.title("Diagnosis ile En YÃ¼ksek Korelasyona Sahip 10 Ã–zellik", fontsize=14)
plt.xlabel("Korelasyon KatsayÄ±sÄ±")
plt.ylabel("Ã–zellik")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Forgetfulness', hue='Diagnosis', data=df, palette='Set2')

plt.title('Forgetfulness ve Diagnosis ArasÄ±ndaki Ä°liÅŸki', fontsize=14)
plt.xlabel('Forgetfulness (0: Yok, 1: Var)', fontsize=12)
plt.ylabel('KiÅŸi SayÄ±sÄ±', fontsize=12)
plt.legend(title='Diagnosis (0: SaÄŸlÄ±klÄ±, 1: Alzheimer)')
plt.show()
```
